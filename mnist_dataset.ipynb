```python
import torch
import torch.nn as nn
import torchvision.datasets as datasets
import torchvision.transforms as transforms
from torch.autograd import Variable


# THIS PROJECT WILL BE ON MNIST DATASET (VERY POPULAR DATASET FOR VISION NN)
```

```python
input_size = 784 # since every image is 28*28 pixels and every pixel is a feature !
output_size = 10 # 0 to 9, 10 digits !
hidden_size = 400 # there are many principles to choose this, two most popular is mean(input,output) or ((2/3)*input + output) !!
epochs = 10
batch_size = 100 # since mnist is large dataset we consider a batch here !
learning_rate = 0.001
```

```python
train_data = datasets.MNIST(root = './mnist', # its where all the images gets downloaded !
                            train=True, # to download training set !
                            transform=transforms.ToTensor(), # to convert images to tensor !
                            download=True) # to downlaod

test_data = datasets.MNIST(root='./mnist',
                           train=False,
                           transform=transforms.ToTensor())
```

```python
# since we want to create many batches we will use dataloader !!
train_loader = torch.utils.data.DataLoader(dataset = train_data,
                                           batch_size = batch_size,
                                           shuffle=True)

test_loader = torch.utils.data.DataLoader(dataset = test_data,
                                          batch_size = batch_size,
                                          shuffle = True)
```

```python
class Net(nn.Module) :
    def __init__(self,input_size,hidden_size,output_size) :
        super(Net,self).__init__()
        self.fc1 = nn.Linear(input_size,hidden_size) # 1st layer
        self.relu = nn.ReLU() # activated 1st layer
        self.fc2 = nn.Linear(hidden_size,hidden_size) # 2nd layer
        self.fc3 = nn.Linear(hidden_size,output_size)
    def forward(self,x) : 
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        out = self.relu(out)
        out = self.fc3(out)
        return out
```

```python
net = Net(input_size, hidden_size, output_size)
cuda = torch.cuda.is_available()
if cuda :
    print(torch.cuda.get_device_name())
    print()
    print(torch.cuda.get_device_capability())
    net = net.cuda()
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(net.parameters(),lr=learning_rate)
```

```python
for i,(images,labels) in enumerate(train_loader) :
    print(images.size())
    print(images.view(-1,784).size())
# in the output op this cell you see (100,1,28,28) this means 
# 100 is the batchsize 
# 1 means the image is grayscale (3 means rgb), this work is done by ToTensor() method !!
# (28,28) is the size in pixels of the images !!
```

```python
correct_train = 0 # how many correct predictions our network makes !
total_train = 0 #  
for e in range(epochs) : 
    for i,(images,labels) in enumerate(train_loader) :
        images = Variable(images.view(-1,28*28))
        labels = Variable(labels)
        if cuda : 
            images = images.cuda()
            labels = labels.cuda()
        optimizer.zero_grad() # clears previous grads !
        outputs = net(images) # this will call forward() method of Net class of alternatively we con do net.forward(images) !
        _,preds = torch.max(outputs.data,1)
        total_train += labels.size(0) # or labels.size[0], just same things !!

        if cuda :
            correct_train += (preds.cpu() == labels.cpu()).sum() # sum function is not implemented on a gpu so we convert to cpu !
        else :
            correct_train += (preds==labels).sum()
        loss = criterion(outputs,labels)
        loss.backward()
        optimizer.step()
        if (i+1)%100 == 0 :
            print(f"Epoch [{e+1}/{epochs}], Iteration [{i+1}/{len(train_data)//batch_size}], Training Loss : {loss.item()}, Training Accuracy : {100*correct_train/total_train}")

print("DONE TRAINING :)")
```

```python
# testing the trained model with test_data !
correct = 0
total = 0
for images,labels in test_loader : 
    images = Variable(images.view(-1,28*28))
    if cuda :
        images = images.cuda()
    outputs = net(images)
    _, preds = torch.max(outputs.data,1)
    total += labels.size(0)
    if cuda :
        correct += (preds.cpu() == labels.cpu()).sum()
    else : 
        correct += (preds == labels).sum()
print(f"Final Test Accuracy : {100*correct/total}")
```
